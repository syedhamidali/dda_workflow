```python
%%time
import warnings
warnings.filterwarnings("ignore")
import pyart
import numpy as np
from matplotlib import pyplot as plt
import glob, os, sys
import shutil
%matplotlib inline
```

```python
files = sorted(glob.glob("COW1/*v1_*SUR.nc"))
len(files)
```

```python
# force the if statement below to be True on first run
el = 99999999
basepath = "."

curr_dir_list = []
for f in files:
    new_el = float(f.split("_")[-2].replace("el", ""))
    if new_el < el:
        # store new dir name
        curr_dir = f.split(".")[1]
        curr_dir_list.append(curr_dir)
        # create directory
        os.makedirs(curr_dir, exist_ok=True)
    # store new el
    el = new_el
    # move file
    shutil.copy(f"{basepath}{os.sep}{f}", f"{basepath}{os.sep}{curr_dir}{os.sep}{f.split(os.sep)[-1]}")
```

```python
radar_files = sorted(glob.glob("20220322/*"))
len(radar_files)
```

```python
import xarray as xr
```

```python
def radar_info(file=None):
    ds = xr.open_dataset(file)
    print("Time: {}".format(ds.time.values))
    print("Volume Number: {}".format(ds.volume_number.values))
    print("Fixed Angle: {}".format(ds.fixed_angle.values))
    print("Sweep Number: {}".format(ds.sweep_number.values))
    print("Range: {}".format(ds.range.values.max()/1e3))
    print("Lat: {}, Lon: {}".format(ds.latitude.values, ds.longitude.values))
```

```python
for file in radar_files:
    radar_info(file=file)
```
